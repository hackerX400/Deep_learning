```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

ReLU activation Function

```python
def reLu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-10, 10, 400)
y = reLu(x)

# Plot the ReLU function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU')
plt.title('ReLU (Rectified Linear Unit) Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Sigmoid Activation Function

Leaky Relu

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

```python
x = np.linspace(-10, 10, 400)
y = leaky_relu(x)


plt.figure(figsize=(8, 6))
plt.plot(x, y, label='leaky_ReLU')
plt.title('LeakyR elu Function')
plt.xlabel('x')
plt.ylabel('Leaky_reLu(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10,10,400)
y = sigmoid(x)
plt.figure(figsize = (8,6))
plt.plot(x,y,label = 'sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.show()
```

Tanh Function

```python
def Tanh(x):
    return  2 * sigmoid(2*x) - 1
```

```python
x = np.linspace(-10, 10, 400)
y = Tanh(x)
plt.figure(figsize = (8,6))
plt.plot(x,y,label = "Tanh")
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.grid(True)
plt.show()
```

ArgMax Function

```python
def ArgMax(x):
    return np.where(x > 0.5, 1,0)
```

```python
x = np.random.rand(400)
y = ArgMax(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="ArgMax", linestyle='None', marker='o')
plt.title('ArgMax Function')
plt.xlabel('x')
plt.ylabel('ArgMax(x)')
plt.grid(True)
plt.legend()
plt.show()
```

SoftMax Function

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
x = np.linspace(-2.0, 2.0, 400)
X = np.vstack([x, x + 1, x - 1])
y = softmax(X)
plt.figure(figsize=(8, 6))
for i in range(X.shape[0]):
    plt.plot(x, y[i], label=f'Softmax curve {i+1}')

plt.title('Softmax Function')
plt.xlabel('x')
plt.ylabel('Softmax(x)')
plt.legend()
plt.grid(True)
plt.show()
```
